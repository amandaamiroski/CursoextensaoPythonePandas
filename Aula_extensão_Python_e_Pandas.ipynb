```python
#Para filtrar uma coluna
display(combustiveis_df['Revenda'])
```

```python
#Para filtrar mais de uma coluna
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']]
display(ca_df)
```

```python
#Para exibir somente linha x
display(ca_df.loc[3])
```

```python
#Para exibir somente da linha x até a y
display(ca_df[9:19])
```

```python
#Para criar um df que tenha somente 4 colunas (revenda, municipio, produto e valor de venda) sobre a gasolina
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)
```

```python
#Para mostrar o valor mais alto de venda
display(gas_df['Valor de Venda'].max())
```
