```python
#Utilizando o dataframe.loc[] com multiplas condições para filtrar
#Ex: postos que vendem etanol em indaiatuba e o preço
#.sort_values serve para mudar a ordem dos valores,
#e o ascending=True é para que seja ordenado de maneira crescente
#ascending=False é para que seja de maneira decrescente

etanol_indaiatuba_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'INDAIATUBA')]
etanol_indaiatuba_df.sort_values(by='Valor de Venda', inplace=True, ascending=True)
display(etanol_indaiatuba_df)
```

```python
#Qual a média de preços dos combustiveis GASOLINA e GASOLINA ADITIVADA do bairro MOOCA, SP?

display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & (combustiveis_df['Municipio'] == 'SAO PAULO') &
                            ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] =='GASOLINA ADITIVADA')), ['Valor de Venda']]. mean())
#Tem que fazer o .loc de cada coluna necessária (Bairro, Minicipio, Produto e Valor de venda)
#Só pode quebrar o código por causa do &
#.mean() foi só para calcular a média
```

```python
#Da para tentar simplificar um pouco usando o insin

display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            (combustiveis_df['Produto'].isin(["GASOLINA", "GASOLINA ADITIVADA"])), 
                            ['Valor de Venda']].mean())
```

```python
# Como mostrar média de valor de venda por combustível de todo o Brasil?
#Como não esta filtrando nada (principalmente linha), somente mostrando uma única coluna não é necessário uso do loc.

media_por_combustivel_df =(ca_df[['Produto', 'Valor de Venda']]).groupby(by='Produto').mean()
display(media_por_combustivel_df)
```

```python
#Para poder inserir novas informações no df
#Quero adicionar uma coluna de valor booleano no combustiveis_df chamada "Ativa" que, por padrão, vai ser true para todas as linhas

combustiveis_df['Ativo'] = True
print(combustiveis_df.info())
display(combustiveis_df.head())

```
