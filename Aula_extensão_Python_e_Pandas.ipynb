```python
#Utilizando o dataframe.loc[] com multiplas condi√ß√µes para filtrar
#Ex: postos que vendem etanol em indaiatuba e o pre√ßo
#.sort_values serve para mudar a ordem dos valores,
#e o ascending=True √© para que seja ordenado de maneira crescente
#ascending=False √© para que seja de maneira decrescente

etanol_indaiatuba_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'INDAIATUBA')]
etanol_indaiatuba_df.sort_values(by='Valor de Venda', inplace=True, ascending=True)
display(etanol_indaiatuba_df)
```

```python
#Qual a m√©dia de pre√ßos dos combustiveis GASOLINA e GASOLINA ADITIVADA do bairro MOOCA, SP?

display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & (combustiveis_df['Municipio'] == 'SAO PAULO') &
                            ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] =='GASOLINA ADITIVADA')), ['Valor de Venda']]. mean())
#Tem que fazer o .loc de cada coluna necess√°ria (Bairro, Minicipio, Produto e Valor de venda)
#S√≥ pode quebrar o c√≥digo por causa do &
#.mean() foi s√≥ para calcular a m√©dia
```

```python
#Da para tentar simplificar um pouco usando o insin

display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            (combustiveis_df['Produto'].isin(["GASOLINA", "GASOLINA ADITIVADA"])), 
                            ['Valor de Venda']].mean())
```

```python
# Como mostrar m√©dia de valor de venda por combust√≠vel de todo o Brasil?
#Como n√£o esta filtrando nada (principalmente linha), somente mostrando uma √∫nica coluna n√£o √© necess√°rio uso do loc.

media_por_combustivel_df =(ca_df[['Produto', 'Valor de Venda']]).groupby(by='Produto').mean()
display(media_por_combustivel_df)
```

```python
#Para poder inserir novas informa√ß√µes no df
#Quero adicionar uma coluna de valor booleano no combustiveis_df chamada "Ativa" que, por padr√£o, vai ser true para todas as linhas

combustiveis_df['Ativo'] = True
print(combustiveis_df.info())
display(combustiveis_df.head())

```

```python
combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
display(combustiveis_df.head())
```

```python
combustiveis_df['Ativo'] = True

display(combustiveis_df.head())
```

```python
combustiveis_df['Obs'] = ["MELHOR CIDADE" if municipio == 'SAO PAULO' else None for municipio in combustiveis_df['Municipio']]
display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['SAO PAULO','INDAIATUBA', 'CAMPINAS', 'SALTO']), ['Municipio', 'Obs']])
```

```python
# como preencher uma coluna 'Status do Valor de Venda' que verifica que se o valor de venda for maior que 6,5 reais,
# ele fala que t√° Caro..caso contr√°rio, est√° barato
import numpy as np

combustiveis_df['Status do Valor de Venda'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Caro', 'Barato')
display(combustiveis_df[['Revenda', 'Valor de Venda', 'Status do Valor de Venda']])
```

```python
#Calcular postos de gasolina por habitantes que temos na amostragem de combust√≠veis de nov/2021
num_habitantes_df = pd.read_csv("ibge_num_habitantes_estimado.csv", sep=";")
num_habitantes_df.rename(columns={"Estado":"Estado - Sigla"}, inplace=True)
display(num_habitantes_df)
```

```python
#Para fazer um merge dos dados
#Onde o inner achar correspondencia ele realiza o merge, se n√£o ele ignora e n√£o adiciona a linha
colunas = ['Municipio', 'Estado - Sigla' ]
merge_df = combustiveis_df.merge(num_habitantes_df, how="inner", on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
#.drop serve para destruir a linha ou coluna que deseja. 
#.dropna destroi a coluna que tem todos os valores nulos
#axis √© para determinar linhas ou colunas, 1-colunas
merge_df.dropna(axis=1, inplace=True)
print(merge_df.info())
```

```python
colunas=['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 'Bairro', 'Cep', 'Produto',
         'Data da Coleta', 'Valor de Venda', 'Unidade de Medida', 'Bandeira', 'Ativo', 'Status do Valor de Venda']
merge_df.drop(labels=colunas, axis=1, inplace=True)
print(merge_df.info())

#o inplace=True serve para salvar como o novo df. Serve como se fosse uma atualiza√ß√£o do df utilizado
```

```python
display(merge_df.head(100))
```

```python
#Para remover as linhas duplicadas
merge_df.drop_duplicates(inplace=True)
display(merge_df.head(100))
```

```python
#Agrupar e contar quantos postos tem na cidade

postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={"Revenda": "N√∫mero de Postos"}, inplace=True)
display(postos_por_municipio_df)

#.groupby serviu para agrupar as informa√ß√µes repetidas
#.count foi para contar o n√∫mero de linhas
#.rename serviu para arrumar o nome das colunas
```
